#### Create dict with most important nouns per story

```python
data['nouns'] = data['nouns'].fillna('')
```

```python
StoryDict = defaultdict(int)

def createDictOfDicts(StoryList):
    for i in range(len(StoryList)):
        #print(i)
        
        # Get story of row
        story = data['STORY'].loc[i]
        
        # Create doc
        if data['nouns'].iloc[200] == data['nouns'].iloc[200]:
            doc = data['nouns'].iloc[i]

            if story in StoryDict:
                # Story has been created within StoryDict
                for word in doc.split():
                    if word in StoryDict[story]:
                        StoryDict[story][word] += 1
                    else:
                        StoryDict[story][word] = 1
            else:
                # Story dict will be created
                for word in doc.split():
                    StoryDict[story] = {}
                    StoryDict[story][word] = 1
```

```python
# New empty dict
MostCommonNounsPerStory = {}

# Get top 3 nouns per story, add to MostCommonNounsPerStory
for story in StoryList1:
    #print(story, StoryDict[story])
    if StoryDict[story] != 0:
        storyCounter = Counter(StoryDict[story])
        MostCommonNounsPerStory[story] = []
        for k,v in storyCounter.most_common(3):
            if k not in MostCommonNounsPerStory[story]:
                MostCommonNounsPerStory[story].append(k)
                for key, value in storyCounter.items():
                    if v == value:
                        if key not in MostCommonNounsPerStory[story]:
                            MostCommonNounsPerStory[story].append(key)
        
```

##### Create column for top 3 nouns per STORY 

```python
data['nounsPerStory'] = data.apply(lambda row: MostCommonNounsPerStory[row.STORY] if row.STORY in MostCommonNounsPerStory.keys() else [], axis=1 )
```

## Create column with location if location is mentioned in Title

```python
dataLoc = data[['TITLE','CATEGORY']]
```

```python
# Function to get locations from TITLE
def preprocessVerb(text):
    # Create Doc object
    doc = nlp(text, disable=['parser'])
    location = [ent.text  for ent in doc.ents if ent.label_ == 'GPE']
    return(location)
  
# Apply preprocess to text
dataLoc['location'] = dataLoc['TITLE'].apply(preprocessVerb)
dataLoc.head()
```
